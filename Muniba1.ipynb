```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
train = pd.read_csv('/content/drive/MyDrive/Muiba_project/Training.csv')
test = pd.read_csv("/content/drive/MyDrive/Muiba_project/Test data.csv")
#check null percentage in each feature column through heatmap
#Age and Cabin features have null values, 
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# train data and labels
train1 = train.drop('Label',axis=1)
train_data = train1.iloc[:1757,:]
test_data = train1.iloc[1757:,:]
train1_label = train['Label']
train1_label[train1_label > 0.8] = 1
train1_label[train1_label < 0.2] = 0
train_label = train1_label.iloc[:1757] 
test_label = train1_label.iloc[1757:] 

#test_data = 
# test data and labels
#test_data = test.drop('Label',axis=1)

#test_label = test['Label']
#test_label[test_label > 0.8] = 1
#test_label[test_label < 0.2] = 0
#test_label
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(train_data,train_label)
```

```python
predictions = logmodel.predict(test_data)
```
